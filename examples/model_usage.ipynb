# Example of using the model module
This notebook shows how to use the components defined in the model. The model module is meant as a light-weight trade model which allows flexible portfolio construction and valuation. The portfolio then handles the position management (long/short), and p&l accounting.

```python
import tia.analysis.model as model
import pandas as pd
```

```python
# Example 
# 1/5/2015, Buy 100 FOO shares at 1
# 1/6/2015, Sell 200 FOO shares at 2
# 1/7/2015, Buy 100 FOO shares at 1.5
#
# This would causes the following
# 1/5/2015: Open Long Position of 100 shares
# 1/6/2015: Close Long Position, Open Short Position of 100 shares
# 1/7/2015: Close Short Position
# 

t1 = model.Trade(1, '1/5/2015', 100., 1., fees=0.3)
t2 = model.Trade(2, '1/6/2015', -200., 2., fees=0.3)
t3 = model.Trade(3, '1/7/2015', 100., 1.5, fees=0.3)
trades = [t1, t2, t3]

# construct a portfolio from the trades
# -create a portfolio pricer (pricer allows for dividends and security multiplier for options or future contracts)
# -create the portfolio
#
pxs = pd.Series([1., 2., 3., 4.], index=pd.date_range('1/5/2015', periods=4))
pricer = model.PortfolioPricer(multiplier=1., closing_pxs=pxs, dvds=None)
port = model.SingleAssetPortfolio(pricer, trades)
```

```python
# show the transaction level frame (takes care of long/short accounting)
port.txns.frame
```

```python
# show the position frame  (column roii is the 'return on initial investment')
port.positions.summary
```

```python
# short the transaction level profit and loss - this includes all days defined by portfolio pricer
port.ltd_txn_pl_frame
```
